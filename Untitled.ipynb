```python
# Prepare transactions for Apriori
transactions = inactive_df.groupby('user_id')['product_name'].apply(list).tolist()
```

```python
# Debugging Step: Check the number of transactions
print(f"Number of transactions: {len(transactions)}")
```

```python
# If no transactions are found, print a message and exit
if len(transactions) == 0:
    raise ValueError("No transactions available for inactive users. Check your data and thresholds.")
```

```python
# Apply Apriori algorithm
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_te = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Create frequent itemsets
frequent_itemsets = apriori(df_te, min_support=0.005, use_colnames=True, low_memory=True)
```

```python
# Debugging Step: Check the frequent itemsets
print(f"Frequent itemsets:\n{frequent_itemsets}")
```

```python
# Check if frequent itemsets are generated
if frequent_itemsets.empty:
    raise ValueError("No frequent itemsets found. Consider lowering min_support or check your data.")
    
```
